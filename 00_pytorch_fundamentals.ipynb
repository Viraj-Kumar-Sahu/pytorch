### pytorch and numpy

```python
arr = np.arange(1., 8.)
tensor = torch.from_numpy(arr).type(torch.float32)
```

##### Note: _PyTorch's_ default dtype is `float32`, whereas _NumPy's_ default dtype is `float64`, so using `torch.from_numpy()` also makes the dtype `float64` use `.type(torch.float32)` to change back to default `float32`

```python
arr.dtype, tensor.dtype
```

```python
tensor = torch.ones(7)
numpy_arr = tensor.numpy()
```

```python
tensor, numpy_arr
```

```python
tensor.dtype, numpy_arr.dtype
```

##### Note: While changing _**NumPy**_ to _**PyTorch**_ (`torch.from_numpy(arr)`) or _**PyTorch**_ to _**NumPy**_ (`tensor.numpy()`), both the operations returns the _new items_ in the _memory_ with _new memory address_.

### Reproducibility (trying to taking out random out of random)

```python
A = torch.rand(size=(3, 4))
B = torch.rand(size=(3, 4))

print(A)
print(B)
print(A == B)
```

```python
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
C = torch.rand(3, 4)
torch.manual_seed(RANDOM_SEED)
D = torch.rand(3, 4)

print(C)
print(D)
print(C == D)
```
